## Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline

sns.set_style('darkgrid')
```

#### Lets read in the dataframe that has all the California Hiking trails with the features and reviewers and ratings from  02_Scrapping_Reviews.ipynb

```python
df = pd.read_csv('../data/california_hikes_df.csv')
```

## 1. Clean Up

#### Check remaining reviewers rating to be collected and null values

```python
df.isnull().sum()
```

#### To check the null in the distance column, it is also the same one for the others.

```python
df[df.distance.isnull()]
```

#### This code will drop the row with the index number 6214 because it is the incorrect trail and website. That is the reason why it has null values for the features.

```python
df.drop(index=6214, inplace=True)
```

#### This code cleans up the distance and elevation columns, strip() removes the white spaces replace() replaces the given string with the second given string in our case it is nothing and converts the object columns to numerical

```python
df['distance'] = df['distance'].map(lambda x: x.strip()).replace(' km\n', '')
df['distance'] = df['distance'].map(lambda x: x.replace(' km', ''))

df['elevation'] = df['elevation'].map(lambda x: x.strip()).replace(' m\n', '')
df['elevation'] = df['elevation'].map(lambda x: x.replace(' m', ''))

df.elevation = df.elevation.map(lambda x: x.replace(',', ''))
df.numreviews = df.numreviews.map(lambda x: x.replace(',', ''))

df.numreviews = df.numreviews.map(lambda x: x.strip())
df.location = df.location.map(lambda x: x.strip())
df.name = df.name.map(lambda x: x.strip())

df.distance = df.distance.astype(float)
df.elevation = df.elevation.astype(int)
df.numreviews = df.numreviews.astype(int)
df.rating = df.rating.astype(float)
```

```python
df.route.value_counts()
```

# 2. Exploration

## a. Distance Exploration

```python
df.distance.describe()
```

The box plot below quickly show the distance of each trail. However, because there is one outlier, it is difficult to visualize the data for distance

```python
sns.boxplot(df.distance);
```

Lets **zoom** in by filtering distances less than 50 miles. <br> Now you can the majority of the trails are less than 10 miles <br> And there a some outside the 75% quartile

```python
fig, ax = plt.subplots(figsize=(14, 7))
sns.boxplot(df[df.distance <= 50].distance);
plt.title('Boxplot of the Distance of the Trails', fontsize=20)
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);
```

This is the number of trails less than or equal to 10 miles

```python
df[df.distance <= 10].distance.count()
```

```python
print(f'Percent less than or equal to 10 miles: {df[df.distance <= 10].distance.count() / df.shape[0]}')
```

### Histogram. Count distribution of the distances of the trails.

```python
fig, ax = plt.subplots(figsize=(14, 7))
plt.hist(df[df.distance < 30].distance, bins=25);
plt.title('Histogram of the Distance of the Trails', fontsize=20)
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);
```

## b. Elevation Exploration

```python
df.elevation.describe()
```

Histogram. Elevation less than and equal to the standard deviation of 5310. The majority of the hiking trails have lower than 1500 ft elevation

```python
fig, ax = plt.subplots(figsize=(15, 7))
plt.hist(df[df.elevation <= 5310].elevation, bins=30);
plt.title('Histogram of the Elevation of the Trails', fontsize=20)
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);

plt.savefig('../plots/hist_elevation_trails.png')
```

## c. Difficulty Exploration

```python
fig, ax = plt.subplots(figsize=(14, 7))
sns.countplot(data=df.groupby(by='difficulty').count(), x=df.difficulty);
plt.title('Distribution of Difficulty of Trails', fontsize=20)
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);

plt.savefig('../plots/dist_difficulty_of_trails.png')
```

## d. Route Exploration 

```python
df.route.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
sns.countplot(data=df.groupby(by='route').count(), x=df.route)
plt.title('Distribution of the Type of Trails', fontsize=20)
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);

plt.savefig('../plots/dist_type_of_trails.png')
```

## e. Rating Exploration

### Histogram & boxplot. Looks like the majority of the hikes/trails rating is around 4

```python
fig, ax = plt.subplots(figsize=(14, 7))
plt.hist(df.rating, bins=6);
plt.title('Distribution of Ratings', fontsize=20)
plt.yticks(fontsize=25)
plt.xticks(fontsize=24);

plt.savefig('../plots/dist_rating.png')
```

```python
sns.boxplot(x=df.rating);
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);
```

## f. Location Exploration

```python
df.location.value_counts().head(10)
```

```python
location_count = [df.groupby(by='location').count().sort_values(
    by='name', ascending=False).name.head(10)[i] for i in range(10)]

fig, ax = plt.subplots(figsize=(20, 15))
g = sns.barplot(x=location_count, 
            y=df.groupby(by='location').count().sort_values(
                by='name', ascending=False).head(10).index)
plt.ylabel(None)
plt.title('Top 10 Locations', fontsize=40)
plt.yticks(fontsize=30)
plt.xticks(fontsize=30);

plt.savefig('../plots/top_10_locations_of_trails.png')
```

## g. Number of Reviews Exploration

```python
df.numreviews.describe()
```

The plot below is the boxplot to show the distrubition of the number of reviews

```python
fig, ax = plt.subplots(figsize=(14, 7))
sns.boxplot(df[df.numreviews <= 110].numreviews);
plt.xlabel('Number of Reviews per Hiking Trail', fontsize=20)
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);
```

The plot belows shows that most trails have less than 10 reviews

```python
fig, ax = plt.subplots(figsize=(15, 7))
plt.hist(df[(df.numreviews < 32) & (df.numreviews > 0)].numreviews, bins=31);
plt.title('0 < Num Of Reviews < 32', fontsize=25)
plt.yticks(fontsize=30)
plt.xticks(fontsize=24);
# less than 32 because 32 is the 75% Quartile

plt.savefig('../plots/num_reviews_less_than_32_greater_than_0.png')
```
